```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from nsepy import get_history
plt.style.use('fivethirtyeight')
```

```python
nifty_50_company_list = ['ADANIPORTS', 'ASIANPAINT', 'AXISBANK','BPCL', 'BAJAJ-AUTO', 'BAJAJFINSV', 'BAJFINANCE','BHARTIARTL', 'CIPLA', 'COALINDIA', 'DRREDDY', 'EICHERMOT', 'GAIL', 
                         'GRASIM', 'HCLTECH', 'HDFCBANK', 'HEROMOTOCO', 'HINDALCO', 'HINDPETRO', 'HINDUNILVR', 'HDFC', 'ITC', 'ICICIBANK', 'IBULHSGFIN', 'IOC', 'INDUSINDBK', 'INFY', 
                         'JSWSTEEL', 'KOTAKBANK', 'LT', 'M&M', 'MARUTI', 'NTPC', 'ONGC', 'POWERGRID', 'RELIANCE', 'SBIN', 'SUNPHARMA', 'TCS', 'TATAMOTORS', 'TATASTEEL', 'TECHM', 
                         'TITAN', 'UPL', 'ULTRACEMCO', 'VEDL', 'WIPRO', 'YESBANK', 'ZEEL', 'ASHOKLEY']
```

```python
result_cols = ['Company_Name','Total_Trade','Positive_Trade','Positive_Trade_%','Negative_Trade','Negative_Trade_%']
result_new_dataframe = pd.DataFrame(columns = result_cols)
print(result_new_dataframe)
```

```python
data = []
total_trade_data = []
no_of_positive_data = []
no_of_negative_data = []
positive_percentage = []
negative_percentage = []
result_data = []
```

```python
for index, company_name in enumerate(nifty_50_company_list):
    
    data.append('data_'+str(index+1))
    total_trade_data.append('total_trade_data_'+str(index+1))
    no_of_positive_data.append('no_of_positive_data_'+str(index+1))
    no_of_negative_data.append('no_of_negative_data_'+str(index+1))
    positive_percentage.append('positive_percentage_'+str(index+1))
    negative_percentage.append('negative_percentage_'+str(index+1))
    result_data.append('result_data_'+str(index+1))

    start = datetime(2020, 1, 1)
    end = datetime(2021, 12, 3)

    data[index] = get_history(symbol=company_name, start=start, end=end)
    total_trade_data[index] = len(data[index])
    
    data[index]['Day Result'] = (data[index]['Close']- data[index]['Open']).round(decimals = 2)
    
    no_of_positive_data[index] = data[index].loc[(data[index]["Day Result"] > 0)]
    no_of_negative_data[index] = data[index].loc[(data[index]["Day Result"] < 0)]
    
    positive_percentage[index] = (len(no_of_positive_data[index]) * 100) / total_trade_data[index]
    negative_percentage[index] = (len(no_of_negative_data[index]) * 100) / total_trade_data[index]
    
    result_data[index] = {'Company_Name': company_name,
         'Total_Trade': total_trade_data[index],
         'Positive_Trade': len(no_of_positive_data[index]),
         'Positive_Trade_%': positive_percentage[index],
         'Negative_Trade': len(no_of_negative_data[index]),
         'Negative_Trade_%': negative_percentage[index]
        }
    result_new_dataframe = result_new_dataframe.append(result_data[index], ignore_index=True)
    
    print(index+1, company_name, positive_percentage[index], '(+)', negative_percentage[index], '(-)')
```
