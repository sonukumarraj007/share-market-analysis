```python
# data = pd.read_csv('temp.csv', parse_dates=["Date"], index_col="Date")
data = pd.read_csv('temp.csv', parse_dates=["Date"])
```

```python
drop_col = ['Symbol', 'Series', 'Trades', 'Turnover']
```

```python
data.drop(drop_col, axis='columns', inplace=True)
```

```python
data['Open Diff']=data['Open']-data['Prev Close']
```

```python
data['Day Close Diff']=data['Close']-data['Open']
```

```python
data['Day Of Week'] = data['Date'].dt.day_name()
```

```python
data['%Deliverble'] = data['%Deliverble']  * 100
```

```python
data.head()
```

```python
# data.reset_index(inplace = True)
data.set_index('Date', inplace=True)
```

```python
data = data[['Day Of Week','Prev Close','Open', 'High', 'Low', 'Last', 'Close','Open Diff', 'Day Close Diff']]
```

```python
data.head()
```

```python
# df['Close'].mean()
data_2010 = data.loc['2010']
data_2011 = data.loc['2011']
data_2012 = data.loc['2012']
data_2013 = data.loc['2013']
data_2014 = data.loc['2014']
data_2015 = data.loc['2015']
data_2016 = data.loc['2016']
data_2017 = data.loc['2017']
data_2018 = data.loc['2018']
data_2019 = data.loc['2019']
data_2020 = data.loc['2020']
data_2021 = data.loc['2021']
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter(company_name+'.xlsx', engine='xlsxwriter')
```

```python
# Write each dataframe to a different worksheet.
data_2010.to_excel(writer, sheet_name='2010')
data_2011.to_excel(writer, sheet_name='2011')
data_2012.to_excel(writer, sheet_name='2012')
data_2013.to_excel(writer, sheet_name='2013')
data_2014.to_excel(writer, sheet_name='2014')
data_2015.to_excel(writer, sheet_name='2015')
data_2016.to_excel(writer, sheet_name='2016')
data_2017.to_excel(writer, sheet_name='2017')
data_2018.to_excel(writer, sheet_name='2018')
data_2019.to_excel(writer, sheet_name='2019')
data_2020.to_excel(writer, sheet_name='2020')
data_2021.to_excel(writer, sheet_name='2021')
```

```python
# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
