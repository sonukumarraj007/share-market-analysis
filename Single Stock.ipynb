```python
import pandas as pd
from nsepy import get_history
from datetime import date
import datetime as dt
```

```python
company_name = 'PNB'
```

```python
data = get_history(symbol=company_name, start=date(2021,5,3), end=date(2021,6,11))
```

```python
drop_col = ['Series', 'Trades', 'Turnover']
```

```python
data.drop(drop_col, axis='columns', inplace=True)
```

```python
data['Open Diff']=data['Open']-data['Prev Close']
```

```python
data['Day Close Diff']=data['Close']-data['Open']
```

```python
data['%Deliverble'] = data['%Deliverble']  * 100
```

```python
data['Volume'] = ((data['Volume'] / 1000000).apply(lambda x: '{:,.1f} M'.format(x)))
```

```python
data['Deliverable Volume'] = (data['Deliverable Volume'] / 1000000).round(decimals=1).astype(str) + ' M'
```
